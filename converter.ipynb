```python
import pandas as pd
from json import load
from html2image import Html2Image
from PIL import Image


data = load(open("TimeTable.json"))

indices = list(data.keys())
timings = list(data[indices[0]].keys())

time_dict = {time:[data[idx][time] if data[idx][time] != None else "" for idx in indices] for time in timings}


df = pd.DataFrame(time_dict)

df["Days"] = indices
df.set_index("Days", inplace=True)

df.to_html("index.html")

hti = Html2Image()

css = r"body {background: white;}"
html = open("index.html", 'r').read()


hti.screenshot(html_str=html, css_str=css, save_as="ttable.png")

print("COMPLETED")
```
